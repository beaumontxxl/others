```python
import requests
import json
import time
```

```python
def get_page(page = 0):
    params = {
        "employer":{"id":"3036416",
                    "name":"Департамент Ф53",
                    "url":"https://api.hh.ru/employers/3036416",
                    "alternate_url":"https://hh.ru/employer/3036416"},
        "page": page,
        "per_page": 100
    }
    
    req = requests.get('https://api.hh.ru/vacancies/', params)
    data = req.content.decode()
    req.close()
    return data
```

```python
js_objs = []
for page in range(0, 20):
    js_objs = json.loads(get_page(page))
    js_objs.append(js_objs["items"])
    if (js_objs['pages'] - page) <= 1:
        break
    time.sleep(0.25)
print('end')
```
